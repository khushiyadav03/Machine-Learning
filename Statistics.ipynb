```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Ek Sample Dataset Banana
# Hum employees ka ek chota sa dataset banate hain
data = {
    'Experience_Yrs': [1, 2, 3, 5, 8, 10, 12, 15],
    'Projects_Completed': [2, 3, 5, 8, 12, 15, 18, 22],
    'Salary_Lakhs': [3.5, 4.5, 6, 9, 14, 18, 22, 28],
    'Mistakes_Monthly': [8, 7, 6, 5, 3, 2, 2, 1]
}
```

```python
# Dictionary se Pandas DataFrame banate hain
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
# .describe() function sabse powerful hai, yeh ek baar me sab kuch de deta hai
print(df.describe())
```

```python
# Measures of Central Tendency
mean_salary = df['Salary_Lakhs'].mean()
median_salary = df['Salary_Lakhs'].median()
mode_projects = df['Projects_Completed'].mode()[0] # .mode() series return karta hai

print(f"Mean Salary: {mean_salary:.2f} Lakhs")
print(f"Median Salary: {median_salary:.2f} Lakhs")
print(f"Mode of Projects Completed: {mode_projects}\n")
```

```python
# Measures of Variability (Data ka Spread)
std_salary = df['Salary_Lakhs'].std()
salary_range = df['Salary_Lakhs'].max() - df['Salary_Lakhs'].min()

print(f"Standard Deviation of Salary: {std_salary:.2f} Lakhs")
print(f"Range of Salary: {salary_range:.2f} Lakhs")
```

```python
# Inferential Statistics (Correlation)

# Pehle correlation matrix calculate karte hain
correlation_matrix = df.corr()

print(correlation_matrix)
```

```python
# Ab is matrix ko visualize karte hain ek Heatmap se
plt.figure(figsize=(8, 6)) # Plot ka size set karte hain
sns.heatmap(
    correlation_matrix,
    annot=True,          # Har cell me value likhega
    cmap='coolwarm',     # Color scheme
    fmt='.2f'            # Values ko 2 decimal places tak format karega
)
plt.title('Feature Correlation Heatmap')
plt.show()
```
