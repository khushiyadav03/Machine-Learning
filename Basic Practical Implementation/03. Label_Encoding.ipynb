Label Encoding: coverting the labels into numeric form

```python
# importing the dependecies
import pandas as pd
from sklearn.preprocessing import LabelEncoder # LabelEncoder ka use labels ko numbers mein badalne ke liye
import sklearn.datasets # Datasets load karne ke liye
from sklearn.model_selection import train_test_split # Data ko train aur test set mein split karne ke liye
```

Label Encoding of Breast Cancer Dataset

```python
# loading the data from csv file to pandas dataframe (csv file se data pandas dataframe mein load kar rahe hain)
cancer_data = sklearn.datasets.load_breast_cancer()
```

```python
# loading the data from csv file to pandas dataframe (csv file se data pandas dataframe mein load kar rahe hain)
cancer_df = pd.DataFrame(cancer_data.data, columns=cancer_data.feature_names)

# first 5 rows of the dataframe (dataframe ki pehli 5 rows dekhte hain)
display(cancer_df.head())
```

```python
# finding the count of different labels (alag alag labels ka count dekhte hain)
cancer_df['area error'].value_counts()
```

```python
# load the label encoder function (label encoder function load kar rahe hain)
label_encoder = LabelEncoder()
```

```python
# 'area error' column ko label encode kar rahe hain (area error column ko numbers mein badal rahe hain)
lables = label_encoder.fit_transform(cancer_df['area error'])
```

Label Encoding of iris data

```python
# loading the data from csv file to pandas DataFrame (csv file se data pandas DataFrame mein load kar rahe hain)
iris_data = pd.read_csv('/content/iris_data.csv')
```

```python
iris_data.head() # iris data ki pehli 5 rows dekhte hain
```

```python
iris_data['Species'].value_counts() # 'Species' column mein alag alag species ka count dekhte hain
```

```python
# loading the label encoder (label encoder load kar rahe hain)
label_encoder_1 = LabelEncoder()
```

```python
# 'Species' column ko label encode kar rahe hain (Species column ko numbers mein badal rahe hain)
iris_labels = label_encoder_1.fit_transform(iris_data.Species)
```

```python
# appending the labels to the DataFrame (labels ko DataFrame mein add kar rahe hain, 'target' naam se naya column banega)
iris_data['target'] = iris_labels
```

```python
iris_data.head() # labels add karne ke baad iris data ki pehli 5 rows dekhte hain
```

```python
iris_data['target'].value_counts() # naye 'target' column mein alag alag encoded labels ka count dekhte hain
```
