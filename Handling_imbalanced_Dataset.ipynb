Imbalanced Dataset: A dataset with an unequal class distribution

```python
# importing the dependecies
import numpy as np # NumPy library ko import kiya, yeh numbers aur arrays ke liye hai
import pandas as pd # Pandas library ko import kiya, yeh data tables (DataFrames) ke liye hai
```

```python
# loading the dataset to pandas DataFrame
# Dataset ko pandas DataFrame mein load kar rahe hain
# 'credit_data.csv' file ko read kar rahe hain
# sep=',' se bata rahe hain ki columns comma (,) se alag hain
# on_bad_lines='skip' se galat format wali lines ko skip kar rahe hain taaki error na aaye
credit_card_data = pd.read_csv('/content/credit_data.csv', sep=',', on_bad_lines='skip')
```

```python
# first 5 rows ko dekh rahe hain
credit_card_data.head()
```

```python
# last 5 rows ko dekh rahe hain
credit_card_data.tail()
```

```python
# 'Class' column mein values count kar rahe hain
# 'Class' column mein kitne 0 aur kitne 1 hain, yeh count dikha raha hai
credit_card_data['Class'].value_counts()
```

This is Highly Imbalanced Dataset

0 --> Legit Transactions

1 --> Fraudulent Transactions

```python
# legit (0) aur fraudulent (1) transactions ko alag alag kar rahe hain
# Legit transactions jahan Class 0 hai
legit = credit_card_data[credit_card_data.Class == 0]
# Fraudulent transactions jahan Class 1 hai
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
# Legit transactions ke DataFrame ka shape (rows, columns) print kar rahe hain
print(legit.shape)
```

```python
# Fraudulent transactions ke DataFrame ka shape (rows, columns) print kar rahe hain
print(fraud.shape)
```

Under-sampling

Build a sample dataset containing similar distribution of Legit & Fraudulent transactions

No of Fraudulent Transactions --> 492

```python
# Legit transactions se sample le rahe hain
# 'n = 492' matlab 492 legit transactions randomly select kar rahe hain
legit_sample = legit.sample(n = 492)
```

```python
# Legit sample DataFrame ko print kar rahe hain
# Jo legit transactions sample kiye hain, unhe dikha rahe hain
print(legit_sample)
```

Concatenate the two DataFrames

```python
# Legit sample aur fraud dataFrames ko combine kar rahe hain
# axis=0 matlab rows ke hisaab se joda ja raha hai
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
# Naye dataset ke shuru ke 5 rows dekh rahe hain
# Combined dataset ka head (shuru ke 5 rows) dikha rahe hain
new_dataset.head()
```

```python
# Naye dataset ke aakhri ke 5 rows dekh rahe hain
# Combined dataset ka tail (aakhri ke 5 rows) dikha rahe hain
new_dataset.tail()
```

```python
# Naye dataset mein 'Class' column ki value counts dekh rahe hain
# Combined dataset mein legit (0) aur fraud (1) transactions ka count dekh rahe hain, ab yeh balanced hona chahiye
new_dataset['Class'].value_counts()
```
