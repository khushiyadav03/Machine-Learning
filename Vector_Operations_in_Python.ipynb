```python
# important libraries import kar rahe hain data ko handle karne aur plot banane ke liye
import numpy as np # numerical operations ke liye
import matplotlib.pyplot as plt # plotting ke liye
import seaborn as sns # advanced plotting aur styling ke liye
```

```python
# Plotting style ko seaborn style par set kar rahe hain, taki plots achhe dikhein
sns.set()
```

Plotting a vector

```python
# Ek vector plot kar rahe hain jo (0,0) se shuru hokar (4,5) tak ja raha hai
plt.quiver(0, 0, 4, 5)
# Plot ko display kar rahe hain
plt.show()
```

```python
# Ek vector plot kar rahe hain (0,0) se (4,5) tak
# angles='xy', scale_units='xy', scale=1: yeh parameters vector ko sahi tarah se scale aur orient karne ke liye hain
plt.quiver(0, 0, 4, 5, angles='xy', scale_units='xy', scale=1)
# x-axis ki limit set kar rahe hain -8 se 8 tak
plt.xlim(-8, 8)
# y-axis ki limit set kar rahe hain -8 se 8 tak
plt.ylim(-8,8)
# Plot ko display kar rahe hain
plt.show()
```

```python
# Ek vector plot kar rahe hain (0,0) se (4,5) tak, blue color mein
# angles='xy', scale_units='xy', scale=1: yeh parameters vector ko sahi tarah se scale aur orient karne ke liye hain
# color='b': vector ka color blue set kar rahe hain
plt.quiver(0, 0, 4, 5, angles='xy', scale_units='xy', scale=1, color='b')
# x-axis ki limit set kar rahe hain -8 se 8 tak
plt.xlim(-8, 8)
# y-axis ki limit set kar rahe hain -8 se 8 tak
plt.ylim(-8,8)
# Plot ko display kar rahe hain
plt.show()
```

```python
plt.quiver(0, 0, 4, 5, scale_units='xy', angles= 'xy', scale=1, color='b')
plt.quiver(0, 0, -3, -6, scale_units='xy', angles='xy', scale=1, color='y')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

Addition of two vectors

```python
vec_1 = np.asarray([0, 0, 2, 3])
vec_2 = np.asarray([0, 0, 3, -2])

sum = vec_1 + vec_2
print(sum)
```

```python
plt.quiver(0, 0, 2, 3, scale_units = 'xy', angles='xy', scale=1, color='b')
plt.quiver(0, 0, 3, -2, scale_units='xy', angles='xy', scale=1, color='g')
plt.quiver(0,0,5,1, scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

Subtraction of 2 vectors

```python
vec_1 = np.asarray([0, 0, 2, 3])
vec_2 = np.asarray([0, 0, 3, -2])

sub = vec_1 - vec_2
print(sub)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='b') # pehla vector plot kar rahe hain (0,0) se (2,3) tak, blue color mein
plt.quiver(0,0,3,-2, scale_units='xy', angles='xy', scale=1, color='g') # dusra vector plot kar rahe hain (0,0) se (3,-2) tak, green color mein
plt.quiver(0,0,-1,5, scale_units='xy', angles='xy', scale=1, color='r') # teesra vector plot kar rahe hain (0,0) se (-1,5) tak, red color mein (yeh subtraction ka result hai)
plt.xlim(-8,8) # x-axis ki limit set kar rahe hain
plt.ylim(-8,8) # y-axis ki limit set kar rahe hain
plt.show() # plot display kar rahe hain
```

Multiplying a vector by a scalar

```python
vec_1 = np.asarray([0, 0, 2, 3]) # pehla vector define kar rahe hain
vec_2 = 2 * vec_1 # pehle vector ko scalar 2 se multiply kar rahe hain
print(vec_2) # result print kar rahe hain
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,4,6, scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
vec_1 = np.asarray([0, 0, 2, 3])
vec_2 = -0.5 * vec_1
print(vec_2)
```

```python
plt.quiver(0, 0, 2, 3, scale_units='xy', angles='xy', scale = 1, color='b')
plt.quiver(0, 0, -1, -1.5, scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```
